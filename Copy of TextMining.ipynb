# Load Aj.Ple Custom Dictionary

```
dictionary = pd.read_csv('https://raw.githubusercontent.com/plenoi/CMU_DataScience/master/mydict.csv', header= None)
```

# Aj.Ple Tokenization Function


```
def isThai(chr):
    cVal = ord(chr)
    if(cVal >= 3584 and cVal <= 3711):
        return True
    return False
    
def tokenize(string, length):
    string = ''.join(list(filter(isThai, string)))
    string = string.replace(' ','')
    result = dict() 
    for word in dictionary:
        if len(word) > length:
            s = 0
            while (1):
                s = string.find(word, s)
                if (s == -1):
                    break
                else:
                    s = s + 1
                    result[s] = word
            string = string.replace(word,'฿'*len(word))  
    return [value for (key, value) in sorted(result.items())]
```

# Feature Extraction using TF-IDF
Change straing data to numerical data

```
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(tokenizer=lambda x:x.split(),ngram_range = (2,2), stop_words =["ฉัน","ผม"],
                     min_df=5,max_df=0.7,norm=None,
                     use_idf=False,smooth_idf=False)
cv.fit(train_clean_join)
X = cv.transform(train_clean_join).toarray()
y = df_train[1]
X
```
